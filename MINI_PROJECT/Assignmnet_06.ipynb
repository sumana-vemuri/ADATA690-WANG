```python
#using altair for visualization 
#here we are taking horsepower vs miles_per_galon for visualization and showing the 
#variation for different countries like europe, USA and japan 
#plotting a point graph
```

```python
alt.Chart(car).mark_point().encode(
    x='Horsepower',
    y='Miles_per_Gallon',
    color='Origin',
)
```

```python
#using altair plotting a bar graph
```

```python
alt.Chart(car).mark_bar().encode(
    x=alt.value(0),
    x2='Horsepower',
    y='Miles_per_Gallon',
    color='Origin',
)
```
