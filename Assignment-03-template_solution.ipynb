```python
# load the 2017-2018 school year file "MERGED2017_18_PP.csv" 

#loading data into pandas
dataframe=pd.read_csv("MERGED2017_18_PP.csv")
#dataframe.head(2)
#(Write code here)
```

Read (with naked eyes, not computer) the dictionary and find out the definitions of the following 12 columns and copy the text next to the column name
- UNITID: (definition here) --the school ID          
- INSTNM:    Institution name     
- CITY:       city of the university/school               
- ZIP:        zip code     
- ADM_RATE:   Admission Rate      
- UGDS:        Number of undergraduate students      
- TUITIONFEE_IN:   In state Tuition fees      
- MN_EARN_WNE_P10: Money earned

```python
# reimport the data with above 12 columns (ignore all other variables)
#(Write code here)

#creating a new dataframe df2 to include the above columns mentioned
column_list=['UNITID','INSTNM','CITY','ZIP','ADM_RATE','UGDS','TUITIONFEE_IN','MN_EARN_WNE_P10']
df2=dataframe[column_list]
print("The new dataframe is:")
print(df2)
```

```python
# Find out the 10 colleges that are hardest to get in (top 10)
#(Write code here)
top10 = df2.nlargest(10, "ADM_RATE") 
print("The 10 colleges that are hardest to get in ", top10)
```

```python
# 10. Find out the average in-state tuition of top 10
fees_top10=top10["TUITIONFEE_IN"].mean()
print("The average in-state tuition of top 10", fees_top10)
#(Write code here)
```

```python
# Find out the average earnings of graduates from the top 10
earnings_top10=top10["MN_EARN_WNE_P10"].mean()
print("The average earnings of graduates from the top 10", earnings_top10)
#(Write code here)
```

```python
# Find out the 10 colleges that are easiest to get in (bottom 10)

#(Write code here)

bottom10 = df2.nsmallest(10, "ADM_RATE")
print("Bottom 10 colleges are:", bottom10)
```

```python
# Find out the average in-state tuition of the bottom 10
#(Write code here)
avg_bottom10_fees=bottom10["TUITIONFEE_IN"].mean()
print("The average in-state tuition of the bottom 10 is",avg_bottom10_fees)
```

```python
# Find out the average earnings of graduates from the bottom 10
#(Write code here)
avg_bottom10_earnings=bottom10["MN_EARN_WNE_P10"].mean()
print("The average earnings of graduates from the bottom 10", avg_bottom10_earnings)
```
