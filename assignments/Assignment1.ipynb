```python

b=0
#taking a global variable for storing sum
sum2=0
#taking an empty list to append the values i.e., the 10 numbers into a list nums
list1=[]
#for reading 10 numbers as inputs
while b<10:
    #increment each value by 1 till 10
    b=b+1
    try:
        print('Enter value of number',+b)
        #taking input 
        x=int(input())
        #appending the numbers entered into the list
        list1.append(x)
        print()
        #sum of all 10 numbers entered
        sum2=sum2+int(x)
       
    except:
        #throws exception if entered value is not a number
        print('please enter only numbers')
        continue
        
#printing list
print("list is",list1)

print("the ten numbers are")
for x in range(len(list1)):
    #print("the ten numbers are")
    print(list1[x])


#calculating the mean i.e., sum/10
mean=sum2/10
print("Mean is",+ mean)

#calculating variance
#variance += (average - number) ** 2
variance=sum([((x - mean) ** 2) for x in list1]) / 10
print("variance is", + variance)

#calculating standard deviation
stddev = variance ** 0.5
print("standard deviation is", + stddev)

#max of 10 numbers
maxVal=max(list1)

#minimum of 10 numbers
minVal=min(list1)

#range of 10 numbers
rangeVal=max(list1)-min(list1)

print("maximum of 10 numbers is", +maxVal)
print("minimum of 10 numbers is", +minVal)
print("range of 10 numbers is", +rangeVal)


#CITATION
#https://www.geeksforgeeks.org/python-get-a-list-as-input-from-user/
#https://www.programiz.com/python-programming/while-loop
#https://www.geeksforgeeks.org/program-find-sum-first-n-natural-numbers/
#https://www.programiz.com/python-programming/examples/add-number
# taking a variable 
#https://stackoverflow.com/questions/26885398/appending-to-list-from-while-loop-python-3
#https://www.geeksforgeeks.org/python-program-to-find-largest-number-in-a-list/
#https://stackoverflow.com/questions/7378091/taking-multiple-inputs-from-user-in-python

```
